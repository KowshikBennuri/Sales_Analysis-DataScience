```python
quarterly_report_group
```

__Plotting the Graphs of weekly report, monthly report and quarterly report for Total Units vs Different Groups__

```python
# Convert SQL result to DataFrame
df_weekly_report_group = pd.DataFrame(weekly_report_group, columns=['WeekNumber','Group', 'StartDate', 'EndDate', 'TotalUnits', 'TotalSales'])

df_weekly_report_group['WeekNumber'] = pd.to_numeric(df_weekly_report_group['WeekNumber'])

# Group data by WeekNumber and group, and calculate total units
grouped_data = df_weekly_report_group.groupby(['WeekNumber', 'Group'])['TotalUnits'].sum().reset_index()

# Plot group-wise units for different groups using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_data, x='WeekNumber', y='TotalUnits', hue='Group', palette='bright')
plt.title('Week-wise Units Analysis for Different Groups')
plt.xlabel('Week')
plt.ylabel('Total Units')
plt.legend(title='Group')

# Convert SQL result to DataFrame
df_monthly_report_group = pd.DataFrame(monthly_report_group, columns=['Month','Group', 'StartDate', 'EndDate', 'TotalUnits', 'TotalSales'])

df_monthly_report_group['Month'] = pd.to_numeric(df_monthly_report_group['Month'])

# Group data by Month and group, and calculate total units
grouped_data = df_monthly_report_group.groupby(['Month', 'Group'])['TotalUnits'].sum().reset_index()

# Plot group-wise units for different groups using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Month', y='TotalUnits', hue='Group', palette='tab10')
plt.title('Month-wise Units Analysis for Different Groups')
plt.xlabel('Month')
plt.ylabel('Total Units')
plt.legend(title='Group')

# Convert SQL result to DataFrame
df_quarterly_report_group = pd.DataFrame(quarterly_report_group, columns=['Group', 'TotalUnits', 'TotalSales'])

# Plot group-wise units for different groups using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df_quarterly_report_group, x='Group', y='TotalUnits',palette='Set2',errorbar=None)
plt.title('Quarterly Units Analysis for Different Group')
plt.xlabel('Group')
plt.ylabel('Total Units')

plt.tight_layout()
plt.show()
```

```python
# Pivot table to summarize sales by group
group_sales = sales_analysis.pivot_table(index='Group', values=['Unit', 'Sales'], aggfunc='sum')
group_sales
```

__Plotting the Graphs of weekly report, monthly report and quarterly report for Total Units vs Different States__

```python
# Convert SQL result to DataFrame
df_weekly_report_state = pd.DataFrame(weekly_report_state, columns=['WeekNumber','State', 'StartDate', 'EndDate', 'TotalUnits', 'TotalSales'])

df_weekly_report_state['WeekNumber'] = pd.to_numeric(df_weekly_report_state['WeekNumber'])

# Group data by WeekNumber and state, and calculate total units
grouped_data = df_weekly_report_state.groupby(['WeekNumber', 'State'])['TotalUnits'].sum().reset_index()

# Plot state-wise units for different state using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_data, x='WeekNumber', y='TotalUnits', hue='State', palette='bright')
plt.title('Week-wise Units Analysis for Different States')
plt.xlabel('Week')
plt.ylabel('Total Units')
plt.legend(title='State')

# Convert SQL result to DataFrame
df_monthly_report_state = pd.DataFrame(monthly_report_state, columns=['Month','State', 'StartDate', 'EndDate', 'TotalUnits', 'TotalSales'])

df_monthly_report_state['Month'] = pd.to_numeric(df_monthly_report_state['Month'])

# Group data by state and month, and calculate total units
grouped_data = df_monthly_report_state.groupby(['Month', 'State'])['TotalUnits'].sum().reset_index()

# Plot state-wise sales for different groups using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_data, x='Month', y='TotalUnits', hue='State', palette='tab10')
plt.title('Month-wise Units Analysis for Different States')
plt.xlabel('Month')
plt.ylabel('Total Units')
plt.legend(title='State')

# Convert SQL result to DataFrame
df_quarterly_report_state = pd.DataFrame(quarterly_report_state, columns=['State', 'TotalUnits', 'TotalSales'])

# Plot state-wise units for different states using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=df_quarterly_report_state, x='State', y='TotalUnits',palette='Set2',errorbar=None)
plt.title('Quarterly Units Analysis for Different States')
plt.xlabel('State')
plt.ylabel('Total Units')

plt.tight_layout()
plt.show()
```

__Daily Sales/Units Analysis(Bar Plot)__

```python
# Extract just the date portion from the index and assign it using .loc
df_oct.loc[:,'Date'] = df_oct.index.day
df_nov.loc[:,'Date'] = df_nov.index.day
df_dec.loc[:,'Date'] = df_dec.index.day

sns.set(rc={'figure.figsize':(25,7)})
fig, axes = plt.subplots(1, 3, sharey=True)
bp_oct = sns.barplot(x = 'Date', y='Unit', data=df_oct, ax=axes[0])
bp_nov = sns.barplot(x = 'Date', y='Unit', data=df_nov, ax=axes[1])
bp_dec = sns.barplot(x = 'Date', y='Unit', data=df_dec, ax=axes[2])
bp_oct.set_xlabel('Oct', fontsize=20)  # Set x-label with fontsize
bp_nov.set_xlabel('Nov', fontsize=20)  # Set x-label with fontsize
bp_dec.set_xlabel('Dev', fontsize=20)  # Set x-label with fontsize
bp_oct.set_title('TotalUnits(Oct)', fontsize=25)  # Set title with fontsize
bp_nov.set_title('TotalUnits(Nov)', fontsize=25)  # Set title with fontsize
bp_dec.set_title('TotalUnits(Dec)', fontsize=25)  # Set title with fontsize
bp_oct.set_ylim(1000, 2000)
bp_nov.set_ylim(1000, 2000)
bp_dec.set_ylim(1000, 2000)
```
